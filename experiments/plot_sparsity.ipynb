```julia
using LinearAlgebra
using Plots
```

```julia
DUMP_DIR = joinpath(@__DIR__, "..", "dump", "sparsity")
NNSDP_PATH = joinpath(@__DIR__, "..", "src", "NnSdp.jl")
include(NNSDP_PATH); using .NnSdp
```

```julia
xdims = [3;3;3;3;3;3;3]
ffnet = randomNetwork(xdims)
zdims = ffnet.zdims
N = sum(zdims[1:end-1])
K = ffnet.K

S(k) = sum(ffnet.xdims[1:k])

E1 = E(1, ffnet.zdims)
EK = E(K, ffnet.zdims)
Ea = E(K+1, ffnet.zdims)
Ein = [E1; Ea]
Eout = [E1; EK; Ea]

hots(X) = map(x -> (x != 0), X)

function quickRawZ(beta)
  Z = zeros(N+1, N+1)
  Z += Ein' * ones(size(Ein)[1], size(Ein)[1]) * Ein
  Z += Eout' * ones(size(Eout)[1], size(Eout)[1]) * Eout
  EMa(k) = VecInt([(S(k-1)+1):min(N, S(k+1)+beta); N+1])
  for k = 1:K-1
    EMk = Ec(EMa(k), N+1)
    Z += EMk' * ones(size(EMk)[1], size(EMk)[1]) * EMk
  end
  return hots(Z)
end


function quickEK()
  Z = zeros(N+1, N+1)
  for i = (S(K-1)+1):N
     for j = 1:N+1
       Z[i,j] = 1
     end
  end
  Z = Z + Z'
  return hots(Z)
end

function quickZ(beta)
  p = 1
  for i in 1:ffnet.K
    if S(i+1) + beta >= S(ffnet.K-1); p = i; break end
  end
  EM(k) = VecInt(S(k-1)+1:S(k+1)+beta)
end
```

```julia
beta = 0
plt = spy(quickRawZ(beta),
  marker = (:square, 8.0, "gray"),
  axis=([], true),
)
```

```julia
saveto = joinpath(DUMP_DIR, "Zrawbeta$(beta).png")
savefig(plt, saveto)
```
